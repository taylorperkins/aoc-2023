```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.ticker import (AutoMinorLocator, MultipleLocator)

from part2 import Graph

%matplotlib inline
```

```python
def get_cycle(fp: str):
    with open(fp) as f:
        aoc = f.read()
    
    graph = Graph([
        list(line)
        for line in aoc.splitlines()
    ])

    start = next(graph.find(lambda n: n.value == "S"))
    return pd.DataFrame([n.coord for n in graph.cycle(start)], columns=["x", "y"])
```

```python
cycle_df = get_cycle("input-test.txt")

plt.plot(cycle_df["x"], cycle_df["y"])
plt.scatter(cycle_df["x"], cycle_df["y"])
plt.show()
```

```python
cycle_df = get_cycle("input-test-4.txt")

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# Set axis ranges; by default this will put major ticks every 25.
ax.set_xlim(-1, 20)
ax.set_ylim(-1, 10)

# Change major ticks to show every 20.
ax.xaxis.set_major_locator(MultipleLocator(5))
ax.yaxis.set_major_locator(MultipleLocator(5))

# Change minor ticks to show every 5. (20/4 = 5)
ax.xaxis.set_minor_locator(AutoMinorLocator(5))
ax.yaxis.set_minor_locator(AutoMinorLocator(5))

# Turn grid on for both major and minor ticks and style minor slightly
# differently.
ax.grid(which='major', color='#CCCCCC', linestyle='--')
ax.grid(which='minor', color='#CCCCCC', linestyle=':')

ax.plot(cycle_df["y"], cycle_df["x"])
ax.scatter(cycle_df["y"], cycle_df["x"])

plt.show()
```

```python
cycle_df = get_cycle("input.txt")

fig = plt.figure(figsize=(20, 20))
ax = fig.add_subplot(1, 1, 1)

# Set axis ranges; by default this will put major ticks every 25.
ax.set_xlim(-1, 140)
ax.set_ylim(-1, 140)

# Change major ticks to show every 20.
ax.xaxis.set_major_locator(MultipleLocator(5))
ax.yaxis.set_major_locator(MultipleLocator(5))

# Change minor ticks to show every 5. (20/4 = 5)
ax.xaxis.set_minor_locator(AutoMinorLocator(5))
ax.yaxis.set_minor_locator(AutoMinorLocator(5))

# Turn grid on for both major and minor ticks and style minor slightly
# differently.
ax.grid(which='major', color='#CCCCCC', linestyle='--')
ax.grid(which='minor', color='#CCCCCC', linestyle=':')

ax.plot(cycle_df["y"], cycle_df["x"])
ax.scatter(cycle_df["y"], cycle_df["x"], s=5)

points_df = pd.DataFrame([
    (i, j)
    for i in range(141)
    for j in range(141)
], columns=["x", "y"])

outer = cycle_df.merge(points_df, how='outer', indicator=True)
anti_join = outer[(outer._merge=='right_only')].drop('_merge', axis=1)
ax.scatter(anti_join["y"], anti_join["x"], s=8, c="red")

plt.savefig("dots.png")
plt.show()
```

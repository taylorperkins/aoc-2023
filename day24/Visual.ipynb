```python
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
paths = []
    
with open("./input-test.txt") as f:

    for idx, line in enumerate(f.read().splitlines()):
        pos, vel = line.split(" @ ")
        x, y, z = (int(v.strip()) for v in pos.split(","))
        vx, vy, vz = (int(v.strip()) for v in vel.split(","))
    
        for step in range(100):
            paths.append({
                "id": idx,
                "step": step,
                "x": x + (step*vx),
                "y": y + (step*vy),
                "z": z + (step*vz),
            })
            

# for step in range(8):
#     paths.append({
#         "id": 99,
#         "step": step,
#         "x": 24 + (step*-3),
#         "y": 13 + (step*1),
#         "z": 10 + (step*2),
#     })
```

```python
df = pd.DataFrame(paths)
df.head()
```

```python
df["id"].unique()
```

```python
fig = plt.figure(figsize=(20, 20))
ax = fig.add_subplot(projection='3d')

for _id in df["id"].unique():
    _df = df[df["id"] == _id]
    
    color = "red" if _id == 99 else "blue"

    ax.scatter(_df["x"], _df["y"], _df["z"], color=color)
    ax.plot(_df["x"], _df["y"], _df["z"], color=color)
```
